#Appliying hyperparamter tuninng

```python
param={
    'criterion':['squared_error','friedman_mse','absolute_error'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,10,15,20,25],
    'max_features':['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid=GridSearchCV(regressor,param_grid=param,cv=5,scoring='neg_mean_squared_error')
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
best_model=DecisionTreeRegressor(criterion='friedman_mse',
 max_depth= 3,
 max_features= 'sqrt',
 splitter ='best')
```

```python
grid.best_estimator_
```

```python
best_model.fit(X_train,y_train)
```

```python
best_pred=best_model.predict(X_test)
```

```python
print(r2_score(y_test,best_pred))
print(mean_absolute_error(y_test,best_pred))
print(mean_squared_error(y_test,best_pred))
```
